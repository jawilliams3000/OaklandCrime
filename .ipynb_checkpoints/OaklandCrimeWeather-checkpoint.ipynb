```python
import pandas as pd
import numpy as np
import urllib2
import seaborn as sb
#from collections import Counter
```

```python
crime_df = pd.read_csv('data/OaklandCrimeData.csv', error_bad_lines=False)
crime_df['DateTime'] = pd.to_datetime(crime_df['DateTime'],format='%Y-%m-%d %H:%M:%S')
crime_df['Date'] = crime_df['DateTime'].apply(lambda x:x.date())
crime_df['WeekDay'] = crime_df['DateTime'].apply(lambda x:x.weekday())
crime_df['HourOfDay'] = crime_df['DateTime'].apply(lambda x:x.hour)
crime_df['Day'] = crime_df['DateTime'].apply(lambda x:x.day)
crime_df['Week'] = crime_df['DateTime'].apply(lambda x:x.week)
crime_df['Month'] = crime_df['DateTime'].apply(lambda x:x.month)
crime_df.head()
```

```python
#Trim crime data to 2012-2013 because of data issues
crime_df = crime_df[crime_df.Year >= 2012]
crime_df = crime_df[crime_df.Year <= 2013]
```

```python
crime_df.CrimeType.value_counts()
```

```python
property_crimes = ['Arson','Burglary','Embezzlement','ForgeryCounterfeiting','Fraud','PossesionStolenProperty','Robbery','Theft','Vandalism','VehicleTheft']
personal_crimes = ['AssaultBattery','Carjacking','ChildElderAbuse','DomesticViolence','Homicide','Kidnapping','RapeAttemptedRape']
quality_of_life_crimes = ['CurfewLoitering','DisorderlyConduct','DisturbingThePeace','DrugPossesionSale','DUI','Gambling','IllegalWeaponsPossesion'
                              ,'ObstructionResistingArrest','OtherSexOffenses','Pandering','Prostitution','PublicDrunkenness']
other_crimes = ['BombThreat','BrandishingWeapon','FirearmDischarge','Other','Threats','Traffic','Unknown','Warrant']
```

```python
property_group = crime_df[crime_df['CrimeType'].isin(property_crimes)].groupby('Date')['ID'].agg(len)
personal_group = crime_df[crime_df['CrimeType'].isin(personal_crimes)].groupby('Date')['ID'].agg(len)
quality_group = crime_df[crime_df['CrimeType'].isin(quality_of_life_crimes)].groupby('Date')['ID'].agg(len)
other_group = crime_df[crime_df['CrimeType'].isin(other_crimes)].groupby('Date')['ID'].agg(len)
```

## Crime by Month Analysis ##

```python
property_df = crime_df[crime_df['CrimeType'].isin(property_crimes)]
personal_df = crime_df[crime_df['CrimeType'].isin(personal_crimes)]
quality_df = crime_df[crime_df['CrimeType'].isin(quality_of_life_crimes)]
other_df = crime_df[crime_df['CrimeType'].isin(other_crimes)]
```

```python
monthly_df = pd.DataFrame()
monthly_df['Personal'] = property_df.groupby('Month')['ID'].agg(len)/len(property_df)*100
monthly_df['Property'] = personal_df.groupby('Month')['ID'].agg(len)/len(personal_df)*100
monthly_df['QualityOfLife'] = quality_df.groupby('Month')['ID'].agg(len)/len(quality_df)*100
#monthly_df['Other'] = other_df.groupby('Month')['ID'].agg(len)/len(other_df)*100

ax = monthly_df.plot(kind='bar')

plt.title('',fontsize=20)
plt.xlabel('Month',fontsize=18)
plt.xticks(plt.xticks()[0],['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'],fontsize=15,rotation=0)
plt.ylabel('Percentage of crimes',fontsize=18)
plt.yticks(fontsize=15)
plt.ylim((0,12))
plt.legend(fontsize=12,ncol=3,loc='upper center')
plt.figsize(3,3)
autolabel(ax.patches)
```
