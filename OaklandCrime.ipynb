```python
# create database
c = MongoClient()
db = c.crime
c.database_names()
```

```python
import json
import urllib
from pandas.io.json import json_normalize

# pull first 1000 records
url = "http://oakland.crimespotting.org/crime-data?dstart=2008-01-01&dend=2014-12-31&beat=&format=json&count=1000"

data = json.load(urllib.urlopen(url))
```

```python
# Insert data into MongoDB
#
event_ids = db.events.insert(data)
```
